```python
import calendar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
calender_df = pd.DataFrame(pd.date_range(start='01/01/2015', end='12/31/2016'), columns=['date'])
calender_df['month'] = calender_df.date.dt.month
```

```python
calender_df['month_sin'] = np.sin(2 * np.pi * calender_df['month']/12.0)
calender_df['month_cos'] = np.cos(2 * np.pi * calender_df['month']/12.0)
calender_df['day_sin'] = np.sin(2 * np.pi * calender_df.date.dt.day/calender_df.date.dt.days_in_month)
calender_df['day_cos'] = np.cos(2 * np.pi * calender_df.date.dt.day/calender_df.date.dt.days_in_month)
```

```python
filtered_df = calender_df[calender_df.date<='12/31/2015']
plt.figure(figsize=(6, 6))
plt.scatter(filtered_df.month_sin, filtered_df.month_cos, color='black')
plt.show()
plt.figure(figsize=(6, 6))
plt.scatter(filtered_df.day_sin, filtered_df.day_cos, color='blue')
```
